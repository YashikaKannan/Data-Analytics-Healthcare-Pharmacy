```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load dataset
df = pd.read_csv("/content/Pharma_data_cleaned.csv", encoding='latin1')

# Normalize column names
df.columns = df.columns.str.strip().str.lower()

print("Available columns:", df.columns)

# 1. Product Type Distribution
plt.figure(figsize=(10,5))
sns.countplot(y='producttypename', data=df, order=df['producttypename'].value_counts().index, palette='Set2')
plt.title("Distribution of Product Types")
plt.xlabel("Count")
plt.ylabel("Product Type")
plt.show()

# 2. Proprietary Names Word Cloud
text = " ".join(df['proprietaryname'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Proprietary Names")
plt.show()


```

```python
# 3. Active Ingredient Frequency
plt.figure(figsize=(10,5))
sns.countplot(y='substancename', data=df, order=df['substancename'].value_counts().head(15).index, palette='Set3')
plt.title("Top 15 Active Ingredients")
plt.xlabel("Count")
plt.ylabel("Substance Name")
plt.show()

# 4. Dosage Strength Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['active_numerator_strength'].dropna(), bins=20, kde=True, color='skyblue')
plt.title("Distribution of Dosage Strengths")
plt.xlabel("Strength")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Dosage Form Distribution
plt.figure(figsize=(10,5))
sns.countplot(y='dosageformname', data=df, order=df['dosageformname'].value_counts().index, palette='Set1')
plt.title("Distribution of Dosage Forms")
plt.xlabel("Count")
plt.ylabel("Dosage Form")
plt.show()

# 6. Route of Administration
plt.figure(figsize=(10,5))
sns.countplot(y='routename', data=df, order=df['routename'].value_counts().index, palette='Set2')
plt.title("Distribution of Routes of Administration")
plt.xlabel("Count")
plt.ylabel("Route")
plt.show()

```

```python
# 7. Marketing Category Distribution
plt.figure(figsize=(10,5))
sns.countplot(y='marketingcategoryname', data=df, order=df['marketingcategoryname'].value_counts().index, palette='Set3')
plt.title("Distribution of Marketing Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()

# 8. Labeler Frequency
plt.figure(figsize=(10,5))
sns.countplot(y='labelername', data=df, order=df['labelername'].value_counts().head(15).index, palette='Set2')
plt.title("Top 15 Labelers")
plt.xlabel("Count")
plt.ylabel("Labeler")
plt.show()
```
